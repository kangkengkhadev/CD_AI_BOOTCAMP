# setup

```python
# เริ่มจากการimportสิ่งที่จพเป็นเข้ามาทำงาน
import numpy as np
import pandas as pd 
```

```python
# โหลดข้อมูลที่เอาไว้เทรนมาใส่ตาราง
df = pd.read_csv('https://raw.githubusercontent.com/kangkengkhadev/CD_AI_BOOTCAMP/main/dataset/moff/train.csv')
```

```python
# เรียกดูข้อมูลคร่าวๆ
df.head()
```

จากโจทย์เราต้องการทำนายสิ่งที่เรียกว่า CO2_working_capacity [mL/g] หรือความสามารถในการกักเก็บco2ของโมเลกุลนั่นเอง โดยที่เราจะใช้ตัวที่เหลือในการทำนาย


```python
# แต่ๆ Mofnameมันเป็นแค่ชื่อลำดับโมเลกุลไม่ได้มีความสัมพันธ์กับคำตอบ เราก็dropมันทิ้งไป
df.drop(['MOFname'],axis=1,inplace=True)
```

```python
df
```

จะเห็นว่าบางค่ามีทั้งnan และ inf ซึ่งนำมาคำนวนไม่ได้ เราจึงต้องลบแถวที่มีค่าพวกนี้ทิ้งไป

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True) # เราก็จะเปลี่ยนให้ inf ให้กลายเป็นnanก่อนจะได้ลบออกง่ายๆ
```

```python
df = df.dropna() # dropแถวที่มีค่าเป็นnanออก
```

```python
df # เรียกดูอีกที
```

```python
df.info() # เช็คข้อมูลของ Dataframe โดยการใช้ฟังก์ชัน .info() จะดูได้ว่ายังมีค่า nan ฟลงเหลืออยู่ไหม
```

```python
col = ["functional_groups", "topology"]
for i in col:
    df[i] = df[i].astype("category").cat.codes # แปลงชื่อคลาสต่างๆที่เป็น string ให้อยู่ในรูปของ integer

df.dtypes # เช็คชนิดของข้อมูลใน Dataframe
```

```python
df.head()
```

```python
df = df[df['surface_area [m^2/g]'] != 0]
```

```python
x = df.drop(['CO2_working_capacity [mL/g]'],axis=1) # ทำการ drop คอลัมม์ที่ต้องจะนำไปให้ทำนายเพียงอย่างเดียว
y = df['CO2_working_capacity [mL/g]'] # กำหนดคอลัมม์ที่ต้องการให้เป็นผลลัพท์ในการทำนาย
```

```python
df
```

# Trainmodel

```python
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
```

```python
model = XGBRegressor(n_estimators=100, max_depth=10, eta=0.1, subsample=0.85, colsample_bytree=0.5) # เรียกโมเดลมาเพื่อใช้งาน
model.fit(x.values,y) # เริ่ม Train โมเดล
```

```python
model2 = XGBRegressor(n_estimators=2000) # เรียกโมเดลมาเพื่อใช้งาน
model2.fit(x.values,y) # เริ่ม Train โมเดล
```

# Predict test data

```python
dftest = pd.read_csv('https://raw.githubusercontent.com/kangkengkhadev/CD_AI_BOOTCAMP/main/dataset/moff/test.csv')
```

```python
dftest
```

```python
dftest.drop(['MOFname'],axis=1,inplace=True) # เอาคอลัมม์ที่ไม่ได้ใช้ออก เนื่องจากไม่มีใน Dataframe ของชุดเทรน
```

```python
col = ["functional_groups",'topology']
for i in col:
    dftest[i] = dftest[i].astype("category").cat.codes # แปลงชื่อคลาสต่างๆที่เป็น string ให้อยู่ในรูปของ integer

dftest.dtypes # เช็คชนิดของข้อมูลใน Dataframe
```

```python
dftest
```

```python
pred = model2.predict(dftest.values) # ทำนายค่าจากโมเดลและชุด Test
```

# Test model

```python
from sklearn.metrics import mean_squared_error
```

```python
y_test = pd.read_csv('https://raw.githubusercontent.com/kangkengkhadev/CD_AI_BOOTCAMP/main/dataset/moff/score.csv')
```
